```python
relevant_columns = ['budget','genres','keywords', 'original_title', 'popularity', 'production_companies', 'vote_average', 'cast', 'director']
```

```python
for feature in relevant_columns:
    movie_data[feature] = movie_data[feature].fillna("")
```

```python
movie_data.isnull().sum()/movie_data.shape[0]*100
```

```python
combined_features = str(movie_data['budget']) + ' ' + movie_data['genres'] + ' ' + movie_data['keywords'] + ' ' + movie_data['original_title'] + ' ' + str(movie_data['popularity']) + ' ' + movie_data['production_companies'] + ' ' + str(movie_data['vote_average']) + ' ' + movie_data['cast'] + ' ' + movie_data['director']
```

```python
#converting the text data into feature vectors
vectorizer = TfidfVectorizer()
feature_vectors = vectorizer.fit_transform(combined_features)
print(feature_vectors)
```

## Cosine Similarity

```python
#getting the similarity scores using cosine similarity
similarity = cosine_similarity(feature_vectors)
print(similarity)
```

```python
similarity.shape
```

```python
#getting the movie name from the user
movie_name = input("Your favourite movie name: ")
#creating a list of all the movies given in the dataset
list_of_all_titles = list(movie_data["title"])
#finding the close match for the movie name given by user
find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)
#close_match
close_match = find_close_match[0]
#finding the index of the movie
index_of_movie = movie_data["index"][movie_data.title == close_match].values[0]
#getting a similarity score for the similar movies
similarity_score = list(enumerate(similarity[index_of_movie]))
#sorting the movies based on their similarity score
sorted_score = sorted(similarity_score, key = lambda x:x[1], reverse=True)
for i, movie in enumerate(sorted_score):
    index = movie[0]
    movie_name = movie_data["title"][movie_data.index == index].values[0]
    if i<30:
        print(i, '.', movie_name)
```
