## Data Preprocessing

```python
#encoding the labels as 0--> benign and 1-->malignant
bcancer_data["diagnosis"].replace({"B":0, "M":1}, inplace=True)
```

```python
bcancer_data["diagnosis"].value_counts()
```

```python
X = bcancer_data.drop("diagnosis", axis=1)
Y = bcancer_data["diagnosis"]
```

```python
params = {
        'n_estimators':[10, 20, 50, 100],
        'criterion' :['gini', 'entropy']
    }
gscv = GridSearchCV(RandomForestClassifier(random_state=40), params , cv=5)
gscv.fit(X,Y)

print('Highest score:',  gscv.best_score_,)
print("best_params:", gscv.best_params_)
```

```python
# params2 = {
#         'C':[1, 5, 10]
#     }
# gscv = GridSearchCV(SVC(kernel='linear'), params2 , cv=5)
# gscv.fit(X,Y)

# print('Highest score:',  gscv.best_score_,)
# print("best_params:", gscv.best_params_)
```

```python
params2 = {
    'kernel':['poly', 'rbf', 'sigmoid'],
        'C':[1]
    }
gscv = GridSearchCV(SVC(kernel='poly'), params2 , cv=5)
gscv.fit(X,Y)

print('Highest score:',  gscv.best_score_,)
print("best_params:", gscv.best_params_)
```

```python
# lgscv = GridSearchCV(SVC(kernel='linear'), {'c': [1, 2, 3]} , cv=5)
# gscv.fit(X,Y)

# print('Highest score:',  gscv.best_score_,)
# print("best_params:", gscv.best_params_)
```

```python
params3 = {
        'C':[1, 5, 10, 20],
        "penalty":["l1","l2"]
    }
gscv = GridSearchCV(LogisticRegression(), params3 , cv=5)
gscv.fit(X,Y)

print('Highest score:',  gscv.best_score_,)
print("best_params:", gscv.best_params_)
```

```python
models = [LogisticRegression('C': 1, 'penalty': 'l2'), SVC('C': 1, 'kernel': 'sigmoid'), RandomForestClassifier('criterion': 'entropy', 'n_estimators': 20)]
for model in models:
    score = cross_val_score(model, X, Y, cv=5)
    mean = score/len(score)*100
    print('score of', model , ':', score)
    print("Mean score of", model, ':', round(mean, 2))
    
```
